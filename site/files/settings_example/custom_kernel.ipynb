The default startup script supports loading settings from the file `settings.json` in folder in which the kernel is started.
The settings are passed to `async_kernel.interface.start_kernel_callable_interface` which allows for traitlets style configuration
to be used when starting the interface.

### Example

'settings.json' in the current folder looks like:

```json
{
  "kernel_class": "echo_kernel.Kernel",
  "flags": ["debug"]
}
```
- `kernel_class`: Overrides the kernel by importing the kernel from 'custom_kernel.py' in the current folder
- `debug`: Sets the flag 'debug' which changes to log level to debug.
