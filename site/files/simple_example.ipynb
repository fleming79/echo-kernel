## Simple example

A simple example to demonstrate concurrent modes. 

Where the top line is a comment `# task` or `# thread` instructs the kernel to modify how the execute request is handled. 

Try running all cells at once with `Run->Run all cells` and then clicking the buttons in a order of you choosing.

```python
import ipywidgets as ipw
from aiologic import REvent

from async_kernel import Caller


async def demo():
    %callers
    b = ipw.Button(description="Continue")
    display(b)
    event = REvent()
    b.on_click(lambda _: event.set())  # noqa: B023
    for i in range(1, 3):
        event.clear()
        b.description = f"Continue {i}"
        print(f"Waiting {i}", end="\r")
        await event
    b.close()
    print("\nDone!")
```

```python
await demo()
```

```python
await demo()
```

```python
# task
await demo()
```

```python
# thread
await demo()
```
